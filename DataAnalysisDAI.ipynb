## FUNCIONES GENERALES


### Función de consulta de casos extremos según una categoría dada.

```python
def analisisCategoria(nomCat):
  a = tuple(datos[['App', 'Price']][datos['Category']==nomCat].sort_values('Price').iloc[0])
  b = tuple(datos[['App', 'Installs']][datos['Category']==nomCat].sort_values('Installs',ascending=False).iloc[0])
  c= tuple(datos[['App', 'Reviews']][datos['Category']==nomCat].sort_values('Reviews', ascending=False).iloc[0])
  d= tuple(datos[['App', 'Last Updated']][datos['Category']==nomCat].sort_values('Last Updated').iloc[0])
  e = tuple(datos[['App', 'Rating']][datos['Category']==nomCat].sort_values('Rating', ascending=True).iloc[0])
  f = tuple(datos[['App', 'Rating', 'Installs']][(datos['Category']==nomCat)&(datos['Rating']==5.0)].sort_values('Installs', ascending=False).iloc[0])
  dicData = {'La más descargada': b, 'Más cara':a, 'Mayor # reviews':c, 'Más tiempo sin actualizar':d, 'Peor rating': e, 'Con rating de 5.0 más descargada': f}
  sd = dicData
  return sd
```

### Prueba:

```python
categoria="ART_AND_DESIGN"
tabla_analisis=analisisCategoria(categoria)
print("Información de la categoría "+categoria+'\n')
for key in tabla_analisis.keys():
  print(key+ " : "+str(tabla_analisis[key])+'\n')
```

### 1. Gráfica de pay de la distribución de aplicaciones de paga y gratuitas.

```python
tipoApps=datos.groupby('Type')['Type'].count().plot(kind='pie')
numFree=datos.groupby('Type')['Type'].count()['Free']
numPaid=datos.groupby('Type')['Type'].count()['Paid']
numTotal=numFree+numPaid
resType={"Type":['Free','Paid'],
       "# apps":[numFree,numPaid],
       "Porcentaje":[str((numFree*100/numTotal).round(2))+'%', str((numPaid*100/numTotal).round(2))+'%']
      }
print(pd.DataFrame(resType))


```

### 6. Encontrar la versión de Android a partir de la cual se vuelven obsoletas más aplicaciones.

```python
ver=dict(datos.groupby('Android Ver').agg({'Android Ver':np.size})['Android Ver'])
total=datos.shape[0]
sum=0
for version in ver.keys():
  if sum>=total/2:
    version_obsoleta=version
    break;
  sum+=ver[version]
print(version_obsoleta)
  
```

### 11. Suma total del dinero generado en ventas de aplicaciones de paga.

```python
categorias=list(datos['Category'])
precios=list(datos['Price'])
descargas=list(datos['Installs'])
dinero_gan=[precios[i]*descargas[i] for i in range(0,len(precios))]
dinero_cat={}
for i in range(0,len(categorias)):
  if categorias[i] in dinero_cat:
    dinero_cat[categorias[i]]+=dinero_gan[i]
  else:
    dinero_cat[categorias[i]]=dinero_gan[i]

print(pd.Series(dinero_cat).sort_values(ascending=False))
pd.Series(dinero_cat).sort_values(ascending=False).head(5).plot(kind='barh')
```

### 12. Aplicación más cara por categoría.

```python
categorias=list(datos['Category'])
precios=list(datos['Price'])
apps=list(datos['App'])
app_cat={}
for i in range(0,len(categorias)):
  if not( categorias[i] in app_cat) or  precios[i] > app_cat[categorias[i]][1]:
    app_cat[categorias[i]]=(apps[i],precios[i])

for i in app_cat.keys():
  if app_cat[i][1]==0:
    app_cat[i]=("Todas son gratis",0.0)
print(pd.Series(app_cat))
```
